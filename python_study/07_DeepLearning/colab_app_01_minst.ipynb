```python
from keras.datasets import mnist
from keras.utils import np_utils
# to_categorical 부르기 위함(one hot encoding 하기 위해서)

import numpy as np
import pandas as pd
import sys
import os
import tensorflow as tf
```

```python
# seed 값 설정

seed=42
np.random.seed(seed)    # 무작위 추출
tf.random.set_seed(42)
```

```python
(X_train, Y_class_train), (X_test, Y_class_test) = mnist.load_data()
```

```python
print('학습셋 이미지 수 : %d 개'%(X_train.shape[0]))
print('test셋 이미지 수 : %d 개'%(X_test.shape[0]))
```

```python
# 차원 변환 (2차원 >> 1차원) reshape
X_train = X_train.reshape(X_train.shape[0], 784) # 784 = 28*28

# 데이터 타입 변경 (정수 >> 실수) 0~255 (흑백사진 : 명도)
X_train = X_train.astype('float64')

# 정규화(normailzation) 0~1
X_train = X_train/255
```

```python
X_test = X_test.reshape(X_test.shape[0], 784).astype('float64') / 255
```

```python
# 바이너리화 과정
Y_train=np_utils.to_categorical(Y_class_train, 10)
Y_test = np_utils.to_categorical(Y_class_test, 10)
```
