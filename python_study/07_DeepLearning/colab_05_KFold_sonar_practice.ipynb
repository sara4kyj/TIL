```python
# 심봤다 테스트

from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold

import pandas as pd
import numpy as np 
import tensorflow as tf

# seed값 설정
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# 데이터 입력

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/02_DeepLearning/dataset/sonar.csv', header=None)
```

```python
# 데이터 개괄적 살펴보기
df.info()
```

```python
df.values
```

```python
dataset = df.values

# x = dataset[:, :-1]
# y_obj = dataset[:, -1]

# ValueError: Failed to convert a NumPy array to a Tensor (Unsupported object type float). 해결방안
x = np.asarray(dataset[:, :-1]).astype(np.float32)
y_obj = dataset[:, -1]
```

```python
x, y_obj
```

```python
# 문자열 변환

le = LabelEncoder()
le.fit(y_obj)
y_encoded = le.transform(y_obj)
```

```python
y_encoded
```

```python
# 10개의 파일로 쪼갬

n_fold = 10
seed = 42

skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
```

```python
accuracy = []

# 모델 설정, 컴파일, 실행

for train, test in skf.split(x,y_encoded):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['accuracy'])
    
    model.fit(x[train], y_encoded[train], epochs=100, batch_size=5)
    k_acc = '%.4f'%(model.evaluate(x[test],y_encoded[test])[1])
    accuracy.append(k_acc)
```

```python
# 결과 출력

print('\n %.2f\n fold accuracy: '%n_fold, accuracy)
```

```python
accuracy = []

# 모델 설정, 컴파일, 실행

for train, test in skf.split(x,y_encoded):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    
    model.fit(x[train], y_encoded[train], epochs=100, batch_size=5)
    k_acc = '%.4f'%(model.evaluate(x[test],y_encoded[test])[1])
    accuracy.append(k_acc)
```
