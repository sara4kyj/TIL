```python
# 심봤다 테스트

from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder

import pandas as pd
import numpy as np 
import tensorflow as tf

# seed값 설정
np.random.seed(42)
tf.random.set_seed(42)
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
# 데이터 입력

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/02_DeepLearning/dataset/sonar.csv', header=None)
```

```python
# 데이터 개괄적 살펴보기
df.info()
```

```python
df.values
```

```python
dataset = df.values

# x = dataset[:, :-1]
# y_obj = dataset[:, -1]

# ValueError: Failed to convert a NumPy array to a Tensor (Unsupported object type float). 해결방안
x = np.asarray(dataset[:, :-1]).astype(np.float32)
y_obj = dataset[:, -1]
```

```python
x, y_obj
```

```python
# 문자열 변환

le = LabelEncoder()
le.fit(y_obj)
y_encoded = le.transform(y_obj)
```

```python
y_encoded
```

```python
# 학습 셋과 테스트 셋 구분(train_test_split())

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.3, random_state=42)
```

```python
# 모델 설정 (모델 형태 구성)

model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # 이진분류 : 시그모이드 함수 사용
```

```python
# 모델 컴파일 (모델 환결성정)

model.compile(loss='mean_squared_error',      # 오차함수 : mse
              optimizer = 'adam',
              metrics=['accuracy'])
```

```python
# 모델 실행 (fit)

# model.fit(x,y_encoded, epochs=120, batch_size=5)
model.fit(x_train,y_train, epochs=120, batch_size=5)
```

```python
# 결과 출력 및 평가

model.evaluate(x_test,y_test)
```

```python
print('\n test_Accuracy: %.4f'%(model.evaluate(x_test, y_test)[1]))
```
