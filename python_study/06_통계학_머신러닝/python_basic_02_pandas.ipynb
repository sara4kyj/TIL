#### DataFrame의 columns 생성, 수정, 삭제

- 생성 : Dataframe['새로운컬럼명'] = 값 또는 식
- 수정 : Dataframe['기존컬럼명'] = 값 또는 식
- 삭제 : Dataframe.drop(labels, axis, inplace)

        labels : 삭제할 컬럼명
        axis = 0 : 행삭제
        axis = 1 : 열삭제
        inplace = False : 삭제후 원본 Dataframe 유지, 삭제된 Dataframe 반환(기본설정)
        inplace = True : 삭제후 원본 Dataframe 변경, 반환값 없음

#### Dataframe의 Data 선택, 필터링
- Dataframe[ ] 의 연산자는 Numpy[ ] 와는 다르다.
- Dataframe[ ] 에 입력 가능한 값

    - 컬럼명(또는 컬럼명 리스트)
    - 불린 인덱스
    - 행의 슬라이싱 연산
    
- loc[ , ] : 명칭 기반 인덱싱 연산자
     - 행 : Dataframe의 index (또는 index 리스트)
     - 열 : Dataframe의 컬럼명 (또는 컬럼명 리스트)
- iloc[ , ] : 위치 기반 인덱싱 연산자   
     - 행 : Dataframe의 행위치
     - 열 : Dataframe의 열위치

#### Mising Data

- NaN은 값이 없는 NULL인 경우이다.
- NaN은 sum(), mean() 등의 함수 연산시 제외된다.
- 머신러닝 알고리즘은 NaN 을 처리하지 않는다.
- isna() : NaN 확인
- fillna() : NaN을 다른 값으로 대체
- dropna() : NaN을 제거


```python
df.isna()  # na : not abailable
```

```python
df.isna().sum()
```

```python
df[df.Age.isna()]
```

```python
na = df[df.Age.isna()]
```

```python
na.sum()
```

```python
df.Embarked.isna().sum()  # method chaining
```

```python
df.Embarked.fillna('N')
```

```python
df.Age.isna().sum()
```

```python
df.Age.dropna()
```

```python
df.dropna()
```

> 바로 dropna를 사용할 경우 유의미한 데이터까지 사라지게 되므로
