```python
# get_dummies 사용, 변환

pd.get_dummies(df.Embarked).head()
```

```python
# 수치형 자료 변환
# 정규화(normalization)

# 1) 정규화 함수 구현 (min max scaling)

def normal(data):
    data = (data - data.min())/(data.max() - data.min())
    return data

df_normal = df.copy()
df_normal.head()
```

```python
df_normal.Fare.head() # 변환전
```

```python
df_normal.Fare = normal(df_normal.Fare)
df_normal.head()
```

```python
# 수치형 자료 변환
# 표준화(Standardization)

# 1) 표준화 함수 구현 

def standard(data):
    data = (data - data.mean())/data.std()
    return data

df_standard = df.copy()
df_standard.head()
```

```python
df_standard.Fare.head()
```

```python
df_standard.Fare = standard(df_standard.Fare)
df_standard.head()
```
