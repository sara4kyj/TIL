```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import tree, ensemble
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
middle = pd.read_csv('./data/middle.csv')
```

```python
x = middle.iloc[:, :-1].values
y = middle.iloc[:, -1].values
```

```python
x, x.shape, y, y.shape
```

```python
# 데이터 정규화

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
```

```python
scaler.transform(x)
```

```python
x_regu = scaler.transform(x)
x_regu.shape, type(x)
```

```python
# 랜덤포레스트 분류기

classifier = ensemble.RandomForestClassifier(n_estimators=20, max_depth=3, criterion='gini')
classifier.fit(x_regu, y)
```

```python
x_regu, type(x_regu),x.shape, y, type(y), y.shape
```

```python
# 실제값
expected = y
expected
```

```python
# 예측값
predicted = classifier.predict(x_regu)
predicted
```

```python
print('accuracy_score:',accuracy_score(expected, predicted))
print('confusion_matrix:\n',confusion_matrix(expected, predicted))
```

#### 에이다부스트_AdaBoost

```python
classifier2 = ensemble.AdaBoostClassifier(base_estimator=tree.DecisionTreeClassifier(max_depth=3), n_estimators=20)
# base를 decision tree모델로 설정
```

```python
classifier2.fit(x_regu, y)
```

```python
# 분류기 성능 확인

expected2 = y
predicted2 = classifier2.predict(x_regu)

print(predicted2)
```

```python
print('accuracy_score:',accuracy_score(expected2, predicted2))
print('confusion_matrix:\n',confusion_matrix(expected2, predicted2))
```
