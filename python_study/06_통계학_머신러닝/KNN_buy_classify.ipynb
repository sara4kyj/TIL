```python
buy = pd.read_csv('./data/buy.csv')
```

```python
buy.columns = ['age', 'paycheck', 'purchase'] # age, paycheck : 연속 / purchase : 범주
buy.head()
```

```python
buy_int = np.array([0]*len(buy))
buy_int[buy.purchase == '구매'] = 1
buy_int[buy.purchase == '비구매'] = 0
buy_int
```

```python
buy.purchase = buy_int
buy
```

```python
# 원본 보관
test_buy = buy
test_buy
```

```python
# 정규화

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(buy)
buy = scaler.transform(buy)
buy
```

```python
# 데이터프레임으로 변경
buy = pd.DataFrame(buy, columns = ['age', 'paycheck', 'purchase'])
buy
```

```python
X_test = test_buy.iloc[:,:2].values
y_test = test_buy.iloc[:,2].values
X_test,y_test
```

```python
from sklearn import neighbors

model = neighbors.KNeighborsClassifier() # 수치예측
model.fit(X_test, y_test)
```

```python
# 결정계수 표시

model.score(X_test, y_test)
```

```python
result = model.predict(X_test)
print(result, y_test, sep='\n\n')
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, result))
print(classification_report(y_test, result))
```
