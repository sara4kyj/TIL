### for문과 while문

#### 1부터 5까지 출력하기
- for문

```python
for i in range(1,6):
    print(i)
```

- while문

```python
n=1
while n<=5:
    print(n)
# 무한루프
```

```python
n=1
while n<=5:
    print(n)
    n+=1
# 무한루프
```

### 1부터 10까지 더하기
- for문

```python
total = 0

for i in range(1,11):
    total += i
    
print(total)
```

- while문

```python
total = 0
n = 1

while n<=10:
    total += n
    n+=1

print(total)
```

### while문만 가능한 경우
- 'q'를 입력할 때까지 반복하여 이름 받기

```python
name = input('Enter the number(If you want to end, press q): ')

while name != 'q':
    name = input('Enter the number(If you want to end, press q): ')
```

- break로 반복문 빠져나오기

```python
while True:
    name = input('Enter the number(If you want to end, press q): ')
    if name == 'q':
        break
```

### 연습문제
- 올바른 아이디/비빌번호를 입력할 때까지 아이디와 비밀번호를 입력하는 프로그램 만들기

```python
id = 'deokku'
pwd = '1234'

while True:
    input_id = input('Enter the ID : ')
    input_pw = input('Enter the PWD : ')
    
    if input_id == id and input_pw == pwd:
        break
    print('Enter it again!')
```

- 확장하기</br>
아이디가 잘못되었으면 'Check your ID' 출력</br>
비밀번호가 잘못되었으면 'Check your PW' 출력

```python
id = 'deokku'
pwd = '1234'

while True:
    input_id = input('Enter the ID : ')
    input_pw = input('Enter the PWD : ')
    
    if input_id == id and input_pw == pwd:
        break
    if input_id != id:
        print('Check your ID')
    if input_pw != pwd:
        print('Check your PW')
```

- 사용자가 0을 입력할 때까지 숫자를 입력받아 입력받은 숫자들의 합을 구하는 프로그램을 작성하세요.

```python
total = 0

while True:
    number = int(input('Enter the number : '))
    
    if number == 0:
        break
    else:
        total += number
print(total)
```
