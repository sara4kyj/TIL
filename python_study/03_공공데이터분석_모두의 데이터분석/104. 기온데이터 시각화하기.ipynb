```python
# 1 8    
import csv
import matplotlib.pyplot as plt

f = open('./data/seoul_book.csv')
data = csv.reader(f)
next(data)
jan = [ ]
aug = [ ]

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        elif month == '08':
            aug.append(float(row[-1]))
    
plt.hist([jan,aug], bins=100, color = ['b','r'])
plt.legend(['jan','aug'])
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('./data/seoul_book.csv')
data = csv.reader(f)
next(data)
jan = [ ]
aug = [ ]

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        elif month == '08':
            aug.append(float(row[-1]))
        
plt.boxplot([jan,aug])
plt.show()
```

```python
import csv
f = open('./data/seoul_book.csv')
data = csv.reader(f)
next(data)
month = [ [] for _ in range(12)]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
plt.boxplot(month)
plt.show()
```

```python
import csv
f = open('./data/seoul_book.csv')
data = csv.reader(f)
next(data)

day = [ [] for _ in range(31)]

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '11':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.boxplot(day)    # outlier()  
plt.show()
```

```python
import csv
f = open('./data/seoul_book.csv')
data = csv.reader(f)
next(data)

day = [ [] for _ in range(31)]

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '11':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)      # ggplot  10,  5 / dpi : 1          
plt.boxplot(day, showfliers = False)    # outlier()  
plt.show()
```
