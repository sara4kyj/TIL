```python
import csv
import matplotlib.pyplot as plt

f = open('./data/seoul_book.csv')
data = csv.reader(f)
# data

# for row in data:
#     print(row)

next(data)

high = []    # 최고기온 담기
low = []     # 최저기온 담기

for row in data :
    if row[-1] != '' and row[-2] != '':                # 최고기온(row[-1])과 최저기온(row[-2]) 값이 있다면,
        date = row[0].split('-')                        # 날짜 값을 '-' 문자 기준으로 구분하여 data에 저장
        if 1988 <= int(date[0]) :                       # 1988 년 이후 데이터라면,
            if date[1] == '05' and date[2] == '07' :   # 5월 7일이라면,
                high.append(float(row[-1]))             # high에 최고기온을 float 형태로 저장
                low.append(float(row[-2]))              # low에 최저기온을 float 형태로 저장
                
plt.rc('font', family = 'Malgun Gothic')     # 맑은 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스기호 깨짐 방지
plt.title('내 생일의 기온 변화 그래프')      # 제목 설정
plt.plot(high, 'hotpink', label = 'high')    # high 리스트에 저장된 값을 hotpink 색으로 그리고 레이블을 표시
plt.plot(low, 'skyblue', label = 'low')      # low 리스트에 저장된 값을 skyblue 색으로 그리고 레이블을 표시
plt.legend()                                 # 범례 표시
plt.show()                                   # 그래프 나타내기
```
