```python
#데이터에서 최고 기온을 실수로 변환,, 한 행씩 출력

import csv

f = open('./data/seoul2.csv', 'r', encoding='cp949')
data=csv.reader(f, delimiter=',')
header = next(data)

for row in data:
    row[5] = float(row[5])      # 최고기온을 실수로 변환
    print(row)

f.close()
```

```python
# 최고 기온과 최고 기온이었던 날짜를 찾으세요.

import csv

f = open('./data/seoul2.csv', 'r', encoding='cp949')
data=csv.reader(f, delimiter=',')
header = next(data)

# 초기화
max_temp=-999
max_temp_date=''

# 최고기온을 찾아서 최고기온과 최고기온이었던 날짜 저장하기
for row in data:
    if row[5] == '':
        row[5] = -999          # 누락된 데이터 처리
    row[5] = float(row[5])     # 최고 기온을 실수로 변환 -> 그래야 비교 가능
    if max_temp < row[5]:
        max_temp=row[5]
        max_temp_date=row[6]

print("최고기온 :",max_temp,"\n기온이 최고로 높았던 날 :", max_temp_date)
f.close()
```
