```python
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)
# cv2는 bgr로 이미지를 가져옴

if img is None :
    print('read failed')
    sys.exit()
    
cv2.namedWindow('image')
cv2.imshow('image', img)

# rgb 변환 cvtColor(src, code[, dst[, dstCn]]) -> dst
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)   

plt.figure(figsize=(12,6))
plt.subplot(131),plt.imshow(img),plt.axis('off')
plt.subplot(132),plt.imshow(imgRGB),plt.axis('off')
plt.subplot(133),plt.imshow(imgGray, cmap = 'gray'),plt.axis('off')
# plt.imshow(imgGray, cmap = 'gray')
# plt.axis('off')
# plt는 rgb로 이미지를 가져옴
plt.show()

while True :
    key = cv2.waitKey()
    if key == 27 or key == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import sys
import os
```

```python
img_list = os.listdir('./fig/images/')

# print(img_list)

img_path = [ './fig/images/'+i for i in img_list]

for path in img_path:
    img = cv2.imread(path, cv2.IMREAD_REDUCED_COLOR_2)
    cv2.namedWindow('image')
    cv2.imshow('image', img)
    
    if cv2.waitKey(1000) == 27:
        break
            
cv2.destroyAllWindows()
```

```python
import sys
import glob
import cv2
import os

#list 형태로 반환
img_list = os.listdir('./fig/images/')

img_path = [ './fig/images/'+i for i in img_list]
    

# img_files_glob = glob.glob('fig/images/*.jpg')
# print(img_files_glob)

# for i in img_files:
#     print(i)


# # 전체 화면으로 'image' 창 생성
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN,
                    cv2.WINDOW_FULLSCREEN)

idx = 0

while True :
    img_name = img_path[idx]
        
    img =cv2.imread(img_name, cv2.IMREAD_COLOR)
    
    cv2.imshow('scene', img)
    
    if cv2.waitKey(2000) == 27:
        break
    
    idx += 1
    
    if idx >= len(img_path): 
        idx = 0
    
cv2.destroyAllWindows()
```
