```python
import matplotlib.pyplot as plt

src1 = cv2.imread('./fig/lenna256.bmp', cv2.IMREAD_GRAYSCALE)
#h, w = src1.shape[:2]
src2 = np.zeros_like(src1, dtype = np.uint8)
cv2.circle(src2, (128,128), 100, 200, -1)
cv2.circle(src2, (128,128), 50, 50, -1)

dst1 = cv2.add(src1, src2)
dst2 = cv2.addWeighted(src1, 0.7, src2, 0.3, 0.0)
dst3 = cv2.subtract(src1, src2)   # src1 - src2
dst4 = cv2.absdiff(src1, src2)

# cv2.imshow('src1', src1)
# cv2.imshow('src2', src2)
# cv2.imshow('dst1', dst1)
# cv2.imshow('dst2', dst2)
# cv2.imshow('dst3', dst3)
# cv2.imshow('dst4', dst4)

# cv2.waitKey()
# cv2.destroyAllWindows()

plt.figure(figsize = (12,6))
plt.subplot(241), plt.imshow(src1, cmap = 'gray'), plt.axis('off'), plt.title('src1')
plt.subplot(242), plt.imshow(src2, cmap = 'gray'), plt.axis('off'), plt.title('src2')
plt.subplot(245), plt.imshow(dst1, cmap = 'gray'), plt.axis('off'), plt.title('dst1')
plt.subplot(246), plt.imshow(dst2, cmap = 'gray'), plt.axis('off'), plt.title('dst2')
plt.subplot(247), plt.imshow(dst3, cmap = 'gray'), plt.axis('off'), plt.title('dst3')
plt.subplot(248), plt.imshow(dst4, cmap = 'gray'), plt.axis('off'), plt.title('dst4')

plt.show()
```

```python
src1 = np.zeros((256,256), np.uint8)
cv2.rectangle(src1, (10,10), (127,248), 255, -1)

src2 = np.zeros((256,256), np.uint8)
cv2.circle(src2, (128,128), 100, 255, -1)

dst_bit_and= cv2.bitwise_and(src1, src2)
dst_bit_or= cv2.bitwise_or(src1, src2)
dst_bit_xor= cv2.bitwise_xor(src1, src2)
dst_bit_not= cv2.bitwise_not(src2)


plt.figure(figsize = (12,6))
plt.subplot(241), plt.imshow(src1, cmap = 'gray'), plt.axis('off'), plt.title('src1')
plt.subplot(242), plt.imshow(src2, cmap = 'gray'), plt.axis('off'), plt.title('src2')

plt.subplot(245), plt.axis('off'), plt.imshow(dst_bit_and, 'gray'), plt.title('dst_bit_and')
plt.subplot(246), plt.axis('off'), plt.imshow(dst_bit_or, 'gray'), plt.title('dst_bit_or')
plt.subplot(247), plt.axis('off'), plt.imshow(dst_bit_xor, 'gray'), plt.title('dst_bit_xor')
plt.subplot(248), plt.axis('off'), plt.imshow(dst_bit_not, 'gray'), plt.title('dst_bit_not_src2')

plt.show()
# cv2.waitKey()
# cv2.destroyAllWindows()
```
