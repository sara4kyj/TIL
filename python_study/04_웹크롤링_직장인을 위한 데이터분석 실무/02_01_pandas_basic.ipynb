```python
code_master = pd.read_excel('./files/sample_codemaster.xlsx',
                           header = 0,
                           usecols = 'A:B',
                           skipfooter = 0)

code_master.head()
```

```python
sample_1.head()
```

```python
sample_1_code = pd.merge(left = sample_1,
                         right = code_master,
                         how = 'left',
                         left_on = '국적코드',
                         right_on = '국적코드')
 
sample_1_code
```

```python
sample_2 = pd.read_excel('./files/sample_2.xlsx',
                         header = 1,
                         skipfooter = 2,
                        usecols = 'A:C')
sample_2['기준년월'] = '2019-12'
sample_2_code = pd.merge(left = sample_2,
                        right = code_master,
                        how = 'inner',
                        left_on = '국적코드',
                        right_on = '국적코드')

sample_2_code
```

```python
sample_2_code['test'] = 'A'
sample = sample_1_code.append(sample_2_code,
                             ignore_index = True)

print(sample)
sample_concat = pd.concat([sample_1_code, sample_2_code],
                         ignore_index = True,
                         axis = 0)
sample_concat
```

```python
sample.to_excel('./files/sample_class_practice.xlsx',
               index = False,
               na_rep = 'NaN',
               sheet_name = 'mysheet')
```

```python
with pd.ExcelWriter('./files/multiple_sheet.xlsx') as writer :
    sample.to_excel(writer, sheet_name='sample_ex1')
    sample_1_code.to_excel(writer, sheet_name='sample_1_code',
                           index = False, 
                           na_rep = 'NaN')    
```

## pivot table

```python
sample_pivot = sample.pivot_table(values = '입국객수',
                                  index = '국적명',
                                  columns = '기준년월',
                                  aggfunc = 'sum'
                                 )

sample_pivot
```

```python
#type(sample_pivot)
sample_pivot.plot(kind='bar')

import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic') 
plt.legend() 
plt.xlabel('')
plt.xticks(size=20)
plt.yticks(size=15)

plt.show()
```
