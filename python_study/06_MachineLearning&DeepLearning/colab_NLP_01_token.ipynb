```python
from tensorflow.keras.preprocessing.text import text_to_word_sequence
```

```python
# 전처리할 텍스트 정하기
text = '해보지 않으면 해낼 수 없다'
```

```python
# 해당 텍스트 토큰화
result = text_to_word_sequence(text)
print('\n원문 : \n', text)
print('\n토큰화 : \n', result)
```

```python
# 텍스트 전처리 함수 Tokenizer()호출
from keras.preprocessing.text import Tokenizer
```

```python
# 전처리하려는 세 개의 문장 정하기
docs = ['먼저 텍스트의 각 단어을 나누어 토큰화 합니다.',
        '텍스트의 단어로 토큰화 해야 딥러닝에서 인식됩니다.',
        '토큰화 한 결과는 딥러닝에서 사용할 수 있습니다.']
```

```python
# 토큰화 함수를 이용해 전처리하는 과정
token = Tokenizer()         # 토큰화 함수 지정
token.fit_on_texts(docs)    # 토큰화 함수에 문장 적용하기
```

```python
# 각 옵션에 맞춰 단어의 빈도 수를 계산한 결과 출력
print('\n 단어 카운트 : \n', token.word_counts)
```

```python
# 출력되는 순서는 랜덤
print('\n문장 카운트:', token.document_count)
print('\n각 단어가 몇 개의 문장에 포함되어 있는가: \n',  token.word_docs)
print('\n각 단어에 매겨진 인데스 값: \n', token.word_index)
```
