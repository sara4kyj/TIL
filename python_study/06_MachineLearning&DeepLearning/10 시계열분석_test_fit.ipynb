```python
data.head()
```

```python
data.plot(figsize=(10,5), linewidth=2, fontsize=10)
plt.xlabel('Year', fontsize=10)
```

```python
# 컬럼명 변경하기
data.columns = ['Energy_Production']
data.head()
```

```python
value = data[['Energy_Production']]
value
```

```python
# 이동평균 구하기

value.rolling(60).mean().plot(figsize=(10,5), linewidth=2, fontsize=8)
plt.xlabel('Year', fontsize=8)
```

```python
# 위 그래프의 차분 그래프 그리기

data.diff().plot(figsize=(10,5), linewidth=2, fontsize=8)
plt.xlabel('Year', fontsize=8)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# 위 그래프의 acf, pacf 곡선 그리기 
# 사용할 데이터는 data

plot_acf(data)
plot_pacf(data)
plt.show()
```

```python
diff_1 = data.diff(periods=1).iloc[1:]
diff_1.plot()
```

```python
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```
