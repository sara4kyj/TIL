```python
import pandas as pd
import numpy as np 

names = ['sepal length','sepal width', 'petal length', 'petal width' ]
```

```python
df.features = pd.DataFrame(X, columns=names)
df.features.head()
```

```python
pd.concat([df.features, df.target], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

x = df.features.values
y = df.loc[:, ['target']].values

x,y
```

```python
# 정규화 (비지도학습, fit_transform())

x = StandardScaler().fit_transform(x)
x, x.shape
```

```python
y, y.shape
```

```python
pca = PCA(n_components=2)

pca_trancformed = pca.fit_transform(x)
```

```python
pca_trancformed, pca_trancformed.shape
```

```python
pd.DataFrame(data=pca_trancformed, columns=['pca_1', 'pca_2'])
```

```python
df_pca = pd.DataFrame(data=pca_trancformed, columns=['pca_1', 'pca_2'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

df_final = pd.concat([df_pca, df.target], axis=1)
```

```python
import matplotlib.pyplot as plt
%pylab inline

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('PCA_component_1', fontsize =15)
ax.set_ylabel('PCA_component_2', fontsize =15)
ax.set_title('2 component PCA', fontsize=20)

targets = [0, 1 , 2 ]
colors = ['r','g','b']

for target, color in zip(targets, colors):
    indicesToKeep = df_final['target'] == target
    ax.scatter(df_final.loc[indicesToKeep, 'pca_1'],
               df_final.loc[indicesToKeep, 'pca_2'],
               c = color,
               s = 50)
    
ax.legend(targets)
ax.grid()
```
